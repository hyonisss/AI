## [3] np.ndarray 생성 : np.array()

```python
a = np.array([0, 1, 2, 3])
print(a)
print(a.ndim)
print(a.shape)
print(len(a))
```

## [5] Basic Data Type

>### np.ndarray.dtype

```python
a = np.array([1, 2, 3])
print(a.dtype)

b = np.array([1., 2., 3.])
print(b.dtype)

c = np.array([1+2j, 3+4j, 5+6*1j])
print(c.dtype)

d = np.array([True, False, False, True])
print(d.dtype)
```

## [6] Indexing & Slicing

> ### 기본 Indexing 방법

```python
a = np.arange(10)
print(a)
print(a[0])
print(a[2])
print(a[-1])
```

>### Multidimensional array의 Indexing

```python
a = np.eye(3)

print(a,'\n')
print(a[1, 1],'\n')

a[2, 1] = 10
print(a[2, 1],'\n')   # third line, second column

print(a,'\n')
print(a[1],'\n')
```

>### Slicing

```python
a = np.arange(10)

print(a,'\n')

print(a[2:9:3],'\n') # [start:end:step]

print(a[:4],'\n')

print(a[1:3],'\n')

print(a[::2],'\n')

print(a[3:],'\n')
```

```python
a = np.arange(10)
print(a,'\n')
a[5:] = 10
print(a,'\n')

b = np.arange(5)
print(b,'\n')
a[5:] = b[::-1]
print(a,'\n')
```

## [7] Dimension reduction 연산

>### np.sum(), np.ndarray.sum()

```python
x = np.array([1, 2, 3, 4])

print(np.sum(x))
print(x.sum())
```

>### 특정 축을 기준으로 합계 구하기

```python
x = np.array([[1, 1], [2, 2]])

print(x.sum(axis=0))

print(x.sum(axis=1))
```

>### np.ndarray.min(), np.ndarray.max(), np.ndarray.argmin(), np.ndarray.argmax() 

```python
x = np.array([1, 3, 2])

print(x.min())
print(x.max())

print(x.argmin())
print(x.argmax())
```

```python
x = np.array([[9,2,10,6,1],[5,7,3,4,8]])

print(x.argmin(axis=0))
print(x.argmax(axis=1))
```

>### np.all(), np.any()

```python
print(np.all([True, True, False]))
print(np.any([True, True, False]))
```

>### 통계 관련

```python
x = np.array([1, 2, 3, 1])
y = np.array([[1, 2, 3], [5, 6, 1]])

print(x.mean())

print(np.median(x))
print(np.median(y, axis=-1))

print(x.std())

np.median(x)
```

## [9] Broadcasting

>### 결과 출력용 함수

```python
def Print_ABC(a, b, c):
    print("{}, a.shape = {}".format(a, a.shape))
    print("{}, b.shape = {}".format(b, b.shape))
    print("{}, c.shape = {}".format(c, c.shape))    
```

>### array + scalar

```python
a = np.array([1, 2, 3])
b = 2
c = a + b

Print_ABC(a, np.array(b), c)
```

>### array + array ex1

```python
a = np.array([[1, 2, 3],[4, 5, 6]])
b = np.array([1, 2, 3])
c = a + b

Print_ABC(a, b, c)
```

>### array + array ex2

```python
a = np.array([1, 2, 3, 4])
b = np.array([[1], [2]])
c = a + b

Print_ABC(a, b, c)
```

>### array + array ex3

```python
a = np.array([[1], [2], [3], [4]])
b = np.array([1, 2, 3])
c = a + b

Print_ABC(a, b, c)
```

>### Flattening : np.ndarry.ravel()

>### Reshaping : np.ndarray.reshape
#### ndarray.reshape may return a view (cf help(np.reshape))), or copy
