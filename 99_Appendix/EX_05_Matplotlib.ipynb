### [예제 2] title, label, legend

```python
plt.plot(["Moon","Song","Lew"],[80,70,95])
plt.plot(["Moon","Song","Lew"],[75,100,80])

plt.title("Exam Score")

plt.xlabel('Name')
plt.ylabel('Score')

plt.legend(["Kor", "Eng"])
plt.show()
```

### [예제 3] subplot

```python
name = ["Moon","Song","Lew"]
score = [80,70,95]

plt.figure(figsize=(12, 4))

# plt.subplot(131)
plt.subplot(131)
plt.bar(name, score)

plt.subplot(132)
plt.scatter(name, score)

plt.subplot(133)
plt.plot(name, score)

plt.show()
```

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(2, 3, figsize=(16, 9))
fig.suptitle("Test")
for idx in range(6):
    ax[idx//3][idx%3].set_title("idx : {}".format(idx))
    ax[idx//3][idx%3].plot(x, x**idx)
```
